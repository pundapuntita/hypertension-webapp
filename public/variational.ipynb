```python
def theta_n(x0,x1,i,eta,mu):
    if (x0==x1).all():
        return mu
    else:
       # T0 = 1
       # T1 = (1+np.sqrt(1+4*T0**2))/2
       # return min( T0/T1, 1/(2**n*np.linalg.norm(x1-x0,'fro')) )
       return min(mu, eta/ np.linalg.norm(x1-x0,'fro')  )

def delta_n(x0,x1,i,eta,rho):
    if (x0==x1).all():
        return rho
    else:
      #  T0 = 1
      #  T1 = (1+np.sqrt(1+4*T0**2))/2
      #  return max( -(T0/T1), -(1/(2**n*np.linalg.norm(x1-x0,'fro')) ))
      return max(rho, -1 * eta / np.linalg.norm(x1-x0,'fro')  )


def thetan(x0,x1,n):
    if (x0==x1).all():
        return 0
    else:
        return 1/(2**n*np.linalg.norm(x1-x0,'fro'))

def theta_iBIGSAM(x0,x1,i,alpha,eta):
    if (x0==x1).all():
        return (i-1)/(i+alpha-1)
    else:
        return min((i-1)/(i+alpha-1), eta/ np.linalg.norm(x1-x0,'fro')  )

def theta_method5(x0,x1,i,tau_n,mu_n):
    if (x0==x1).all():
        return mu_n
    else:
        return min((mu_n), tau_n/ np.linalg.norm(x1-x0,'fro')  )

def prox(x,L1,L2,cn):
    r = x - np.dot(L1,x) + L2
    s = np.abs(r)-cn #s=x_i - lambda
    s = np.maximum(s,0)
    return s*np.sign(r)

def T(x,L1,L2,cn):
    r = x - np.dot(L1,x) + L2
    s = np.abs(r)-cn #s=x_i - lambda
    s = np.maximum(s,0)
    return s*np.sign(r)


def prox2(x, cn, lam, L1, L2):
    # Compute the gradient of f(x)
    gradf = 2*( np.dot(L1,x) - L2)
    # Compute y = x - cn * gradf
    y = x - cn * gradf
    # Compute the proximal operator value
    value = np.maximum(np.abs(y) - lam * cn, 0) * np.sign(y)

    return value  # Return the prox_cn g(I - cn * gradf)(x) value


#contraction
def F(x):
    return 0.999*x

def F1(x):
    return (1/3)*x+1


def sum_theta(x,theta, num_in, q):
    value = 0
    for j in range(0,num_in):
        value += theta * np.linalg.norm(x[q-j] - x[q-1-j], ord=2)
    return value

def theta_miBIGSAM(x, i, alpha, eta, q):
    sum_norm = sum_theta(x,1, num_in, q)
    if (sum_norm == 0).all():
        return (i-1)/(i+alpha-1)
    else:
        return min((i-1)/(i+alpha-1), eta/ sum_norm  )

def theta_n2(x0,x1,i,eta,theta_bar):
    if (x0==x1).all():
        return theta_bar
    else:
       return theta_bar/((i**2)* np.linalg.norm(x1-x0,'fro'))

def calculate_metrics(y_true, y_pred, average='binary'):
    precision = precision_score(y_true, y_pred, average=average, zero_division=1)
    recall = recall_score(y_true, y_pred, average=average, zero_division=1)
    f1 = f1_score(y_true, y_pred, average=average, zero_division=1)
    return precision, recall, f1
```

# AIC BIC Function

```python
import numpy as np

def calculate_aic_bic_from_rss(y_true, y_pred, num_params):
    n = len(y_true)
    residuals = y_true - y_pred
    rss = np.sum(residuals ** 2)
    sigma_squared = rss / n
    log_likelihood = -0.5 * n * (np.log(2 * np.pi) + np.log(sigma_squared) + 1)
    aic = 2 * num_params - 2 * log_likelihood
    bic = np.log(n) * num_params - 2 * log_likelihood
    return aic, bic

def compute_aic_bic_lists(pred_list, y_list, num_params):
    aic_list = []
    bic_list = []
    for y_pred, y_true in zip(pred_list, y_list):
        y_pred = np.array(y_pred)
        y_true = np.array(y_true)
        if set(np.unique(y_pred)) - {0, 1}:
            y_pred = (y_pred > 0.5).astype(int)
        aic, bic = calculate_aic_bic_from_rss(y_true, y_pred, num_params)
        aic_list.append(aic)
        bic_list.append(bic)
    return aic_list, bic_list
```

##*3.3 Algorithm*

```python
def Puntita(prox, L1, L2, Lip, m, n, itrs_max):
    x0 = np.zeros((m,n))
    x1 = np.zeros((m,n))
    x2 = np.zeros((m,n))
    lam1 = 1e-5
    for i in range(1,itrs_max+1):
        alpha = 3
        sigma = 1
        mu_n = 1 - 1/(i+1)
        xi_n = 33*(10**20)/(i**2)
        rho = - 1/i
        eta_n = 1/(i+2)
        beta_n = i/(2*i+1)
        tau_n = 0.5 + 1/(2*i)
        alpha_n = 1/(2*i+1)
        #cn = (1/2)*Lip #gamma
        cn = Lip #gamma




        y2 = x2 + theta_n(x2,x1,i,xi_n,mu_n)*(x2-x1) + delta_n(x0,x1,i,xi_n,rho)*(x1-x0)
        z2 = y2 + eta_n*(x2-y2)
        v2 = (1-beta_n)*x2 + beta_n*prox(z2,cn,lam1,L1,L2)
        w2 = (1-tau_n)*prox(y2,cn,lam1,L1,L2) + tau_n*prox(v2,cn,lam1,L1,L2)
        x3 = alpha_n*F1(x2)+(1-alpha_n)*w2

        x0, x1, x2 = x1, x2, x3  # Update values for the next iteration

    Beta = x3  # Final result
    return Beta

```

```python
import numpy as np

def fasta_fp_pp_ultimate_v2_final_safe(prox, L1, L2, Lip, m, n, itrs_max):
    """
    FASTA-FP Ultimate safe version, signature matching Puntita/BiGSAM/iBiGSAM.
    prox: ignored (we use internal l1 prox)
    L1 (H): (num_samples, m) design matrix
    L2 (Y): (num_samples, n) target matrix
    Lip: Lipschitz constant
    m: hidden dimension (n_features)
    n: output dimension (n_outputs)
    itrs_max: maximum iterations
    Returns Beta (m × n)
    """
    # Hyperparameters
    lam = 1e-3       # l1 regularization weight
    eta0 = 0.9       # inertia base
    clip_val = 1e3   # clipping threshold
    tol = 1e-5       # convergence tolerance

    # ensure L2 is 2D
    if L2.ndim == 1:
        L2 = L2.reshape(-1, 1)
    num_samples, num_features = L1.shape
    _, num_outputs = L2.shape

    # initialize iterates x0, x1, x2 shape (m, n)
    x0 = np.zeros((num_features, num_outputs))
    x1 = np.zeros_like(x0)
    x2 = np.zeros_like(x0)

    # gradient step size
    step = 1.0 / Lip

    for k in range(1, itrs_max + 1):
        beta  = 1.0 / (k + 1)
        theta = eta0 * (k - 1) / (k + 2)
        delta = -eta0 / (k + 1)

        # gradient: 2 * H^T (H x1 - Y)
        resid = L1.dot(x1) - L2                    # (num_samples, n)
        grad  = 2 * (L1.T.dot(resid))              # (m, n)
        grad  = np.clip(np.nan_to_num(grad), -clip_val, clip_val)

        # inertial extrapolation
        d10 = np.clip(x1 - x0, -clip_val, clip_val)
        d02 = np.clip(x0 - x2, -clip_val, clip_val)
        y   = x1 + theta * d10 + delta * d02
        y   = np.clip(np.nan_to_num(y), -clip_val, clip_val)

        # gradient descent
        z = y - step * grad
        z = np.clip(np.nan_to_num(z), -clip_val, clip_val)

        # internal l1 proximal (soft-threshold)
        thresh = lam * step
        x_prox = np.sign(z) * np.maximum(np.abs(z) - thresh, 0)
        x_prox = np.clip(np.nan_to_num(x_prox), -clip_val, clip_val)

        # momentum combine
        x_new = (1 - beta) * x1 + beta * x_prox
        x_new = np.clip(np.nan_to_num(x_new), -clip_val, clip_val)

        # check convergence
        if np.linalg.norm(x_new - x1) < tol:
            return x_new

        # shift for next iteration
        x2, x0, x1 = x0, x1, x_new

    return x1

```

```python
def Pronpat(prox, L1, L2, Lip, m, n, itrs_max):
    x0 = np.zeros((m,n))
    x1 = np.zeros((m,n))
    lam1 = 1e-5
    for i in range(1,itrs_max+1):
        alpha = 3
        sigma = 1
        mu_n = 1 - 1/(i+1)
        xi_n = 33*(10**20)/(i**2)
        rho = - 1/i
        eta_n = 1/(2*i+1)
        beta_n = i/(2*i+1)
        tau_n = 0.5 + 1/(2*i)
        alpha_n = 1/(2*i+1)
        #cn = (1/2)*Lip #gamma
        cn = Lip #gamma



        theta_bar = 1/((x1 - x0)**5+i**5)
        y1 = x1 + theta_n2(x0,x1,i,xi_n,theta_bar)*(x1-x0)
        z1 = y1 + eta_n*(x1-y1)
        x2 = prox(z1,cn,lam1,L1,L2)

        x0, x1 = x1, x2  # Update values for the next iteration

    Beta = x2  # Final result
    return Beta
```

```python
def BIGSAM(prox, L1, L2, Lip, m, n, itrs_max):
    x1 = np.zeros((m,n))

    lam1 = 1e-5 #10^-5

    for i in range(1,itrs_max+1):
        cn = Lip
        alpha_n = 1/(n+1)
        eta = 1/(n+1)**2
        y = prox(x1,cn,lam1,L1,L2)
        z = F(y)
        x = alpha_n*z + (1-alpha_n)*y # x_{n+1}    รันรอบแรกจะได้ x_2 ออกมา

        x0, x1 = x1, x      #อัพเดทค่า x ก่อนจะรันรอบต่อไป
    Beta = x
    return Beta
```

```python
def iBIGSAM(prox, L1, L2, Lip, m, n, itrs_max):
    x1 = np.zeros((m,n))
    x0 = np.zeros((m,n))
    lam1 = 1e-5 #10^-5

    for i in range(1,itrs_max+1):
        cn = Lip
        alpha = 3
        alpha_n = 1/(i+1)
        eta = 1/(i+1)**2
        v = x1 + theta_iBIGSAM(x0,x1,i,alpha,eta)*(x0-x1)
        y = prox(v,cn,lam1,L1,L2)
        z = F(v)
        x = alpha_n*z + (1-alpha_n)*y # x_{n+1}    รันรอบแรกจะได้ x_2 ออกมา

        x0, x1 = x1, x      #อัพเดทค่า x ก่อนจะรันรอบต่อไป
    Beta = x
    return Beta
```

```python
def aiBIGSAM(prox, L1, L2, Lip, m, n, itrs_max):
    # Initialization and setup
    x1 = np.zeros((m, n))
    x0 = np.zeros((m, n))
    lam1 = 1e-5  # 10^-5
    for i in range(1, itrs_max + 1):
        cn = Lip
        alpha = 3
        alpha_n = 1 / (i + 1)
        eta = 1 / (i + 1) ** 2

        if i % 2 == 0:
            v = x1
        else:
            v = x1 + theta_iBIGSAM(x0, x1, i, alpha, eta) * (x0 - x1)

        y = prox(v,cn,lam1,L1,L2)
        z = F(v)
        x = alpha_n * z + (1 - alpha_n) * y  # x_{n+1}

        x0, x1 = x1, x  # Update values for the next iteration
    Beta = x  # Final result
    return Beta
```
